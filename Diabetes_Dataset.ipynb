# DIABETES DATASET :- Machine Learning Assignment
This notebook applies multiple machine learning algorithms
on the Diabetes dataset using different train_test splits.
The datasets is converted into classification format and evaluated using accuracy.

```python
#
from sklearn.datasets import load_diabetes
from sklearn.model_selection import  train_test_split
from sklearn.metrics import accuracy_score

from sklearn.naive_bayes import GaussianNB
from sklearn.neural_network import MLPClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier

# Load dataset
data = load_diabetes()
X = data.data

# Convert to classification
y = (data.target > data.target.mean()).astype(int)

models = {
    "Naive Bayes": GaussianNB(),
    "MLP Classifier": MLPClassifier(max_iter=1000),
    "Decision Tree" : DecisionTreeClassifier(),
    "Random Forest": RandomForestClassifier(),
    "SVM": SVC(),
    "KNN ": KNeighborsClassifier()
}

splits = [0.2 , 0.3 , 0.5] # Corrected invalid split values

for split_ratio in splits:
    print(f"\nTrain_test Split : {int((1-split_ratio)*100)} : {int(split_ratio*100)}")

    X_Train, X_test , y_train , y_test = train_test_split(X , y , test_size= split_ratio , random_state=42)

    for name , model in models.items():
        model.fit(X_Train , y_train)
        y_pred = model.predict(X_test)
        acc= accuracy_score(y_test , y_pred)
        print(f"{name} ; {round(acc ,3)}")
```
